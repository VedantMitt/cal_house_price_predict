```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 📍 1. Select only the location features
coords = housing[["latitude", "longitude"]]

# 🔢 2. Apply K-Means (e.g., 5 clusters)
kmeans = KMeans(n_clusters=5, random_state=42)
housing["region_cluster"] = kmeans.fit_predict(coords)

# 🎨 3. Optional: Visualize clusters
plt.figure(figsize=(10, 6))
plt.scatter(housing["longitude"], housing["latitude"], c=housing["region_cluster"], cmap="tab10", alpha=0.6)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Region Clusters Based on Location")
plt.colorbar(label="Cluster")
plt.show()

coords_test = strat_test_set[["latitude", "longitude"]]
strat_test_set["region_cluster"] = kmeans.predict(coords_test)
```

```python
# Separate target and features
housing_features = housing.drop("median_house_value", axis=1)
housing_labels = housing["median_house_value"]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

# 1️⃣ Separate columns

cat_attribs = ["ocean_proximity", "region_cluster"]
num_attribs = list(housing_features.drop(cat_attribs, axis=1))  # all numeric cols

# 2️⃣ Numeric pipeline
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("std_scaler", StandardScaler()),
])

# 3️⃣ Full pipeline (combine num + cat)
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])


# 4️⃣ Apply pipeline to data
housing_prepared = full_pipeline.fit_transform(housing_features)

X_test = strat_test_set.drop("median_house_value", axis=1)
y_test = strat_test_set["median_house_value"]
X_test_prepared = full_pipeline.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Train the model
lin_reg = LinearRegression()
lin_reg.fit(housing_prepared, housing_labels)

# Predict on training data
housing_predictions = lin_reg.predict(housing_prepared)

# Evaluate the model
lin_mse = mean_squared_error(housing_labels, housing_predictions)
lin_rmse = lin_mse ** 0.5

print("Root Mean Squared Error:", lin_rmse)
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(housing_prepared, housing_labels)

housing_predictions = tree_reg.predict(housing_prepared)
tree_mse = mean_squared_error(housing_labels, housing_predictions)
tree_rmse = tree_mse ** 0.5

print("Decision Tree RMSE (train set):", tree_rmse)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

# Perform 10-fold cross-validation (negated MSE is returned)
scores = cross_val_score(tree_reg, housing_prepared, housing_labels,
                         scoring="neg_mean_squared_error", cv=10)

# Convert to RMSE
tree_rmse_scores = np.sqrt(-scores)

# Print results
print("Cross-Validation RMSE Scores:", tree_rmse_scores)
print("Mean:", tree_rmse_scores.mean())
print("Standard Deviation:", tree_rmse_scores.std())
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
import numpy as np

# Define parameter grid
param_grid = {
    'n_estimators': [100, 200],
    'learning_rate': [0.1],
    'max_depth': [3, 5, 7],
    'subsample': [0.8]
}

# Set up GridSearchCV
xgb_reg = XGBRegressor(random_state=42)
grid_search = GridSearchCV(xgb_reg, param_grid, cv=5,
                           scoring="neg_mean_squared_error",
                           return_train_score=True,
                           n_jobs=-1, verbose=1)

# Fit the model
grid_search.fit(housing_prepared, housing_labels)

# Get best model
final_model = grid_search.best_estimator_

# Prepare the test set
X_test = strat_test_set.drop("median_house_value", axis=1)
y_test = strat_test_set["median_house_value"]
X_test_prepared = full_pipeline.transform(X_test)

# Predict and evaluate
final_predictions = final_model.predict(X_test_prepared)
final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)

print("Test Set RMSE:", final_rmse)
```

```python
from sklearn.ensemble import VotingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# Define base models
xgb_model = XGBRegressor(n_estimators=200, learning_rate=0.1, max_depth=7, subsample=0.8, random_state=42)
rf_model = RandomForestRegressor(n_estimators=200, random_state=42)
lin_model = LinearRegression()

# Combine into a Voting Regressor
voting_reg = VotingRegressor(estimators=[
    ('xgb', xgb_model),
    ('rf', rf_model),
    ('lin', lin_model)
], weights=[3, 1, 1])  # XGBoost weighted 3×)

# Fit on training data
voting_reg.fit(housing_prepared, housing_labels)

# Evaluate on test set
final_predictions = voting_reg.predict(X_test_prepared)
final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)

print("Voting Regressor Test Set RMSE:", final_rmse)
```
